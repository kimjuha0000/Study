Date (날짜): 주식 거래가 이루어진 날짜입니다.

Open (시가): 주식의 거래가 시작된 가격입니다.

High (고가): 해당 날짜 동안 기록된 최고 주가입니다.

Low (저가): 해당 날짜 동안 기록된 최저 주가입니다.

Close (종가): 해당 날짜의 거래가 종료된 시점의 주가입니다.

Volume (거래량): 해당 날짜 동안 거래된 주식의 총 수량입니다.

Change (변동률): 전일 종가 대비 종가의 변동률입니다.

```python
plt.figure(figsize = (14,7))
plt.plot(df_samsung['Close'], label = 'samsung')
plt.plot(df_kakao['Close'], label = 'kakao')
plt.legend()
```

```python
df_samsung['Close'].pct_change() #종가에 대한 일별수익률
```

image.png

```python
samsung_pct = (df_samsung['Close'].pct_change()*100)
samsung_pct.describe()
```

```python
plt.figure(figsize=(15,10))
plt.plot(samsung_pct.index, samsung_pct, color = 'b', linestyle = '-')
plt.title('Daily Return of Samsung Electronics')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
```

```python
plt.figure(figsize = (10,6))
plt.hist(samsung_pct, bins = 30, color = 'skyblue')
plt.title('Histogram of Daily Return of Samsung Electronics')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
```

```python
df_samsung['ret']= (df_samsung['Close'].pct_change()*100)
df_samsung['ret']
```

```python
df_samsung['ret'].dropna(inplace = True)
```

```python
len(df_samsung['ret'])
```

```python
positive = 0

n = len(df_samsung['ret'])

for i in range(n):
    if df_samsung['ret'][i] > 0:
        positive += 1

positive
```

```python
negative = n-positive
negative
```

```python
positive_rate = positive/n
positive_rate
```
